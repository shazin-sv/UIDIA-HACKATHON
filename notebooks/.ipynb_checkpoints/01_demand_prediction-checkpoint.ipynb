### 1. Introduction

**Project Setu** is a proactive, life-event-driven analytics layer designed for UIDAI to manage Aadhaar identity updates more efficiently. This notebook focuses on the **Mandatory Biometric Update (MBU)** demand prediction.

#### Mandatory Biometric Update (MBU)
Aadhaar requires mandatory biometric updates for children when they reach the age of **5** and **15**. These updates are critical for maintaining the accuracy and security of the Aadhaar database. 

#### Objective
The goal of this analysis is to forecast the 6-month future demand for MBUs at a **pincode level**. By predicting these surges in advance, UIDAI can:
- **Optimize Infrastructure**: Allocate mobile vans and temporary camps to high-demand areas.
- **Manpower Planning**: Ensure adequate staffing at Aadhaar Seva Kendras (ASKs).
- **Proactive Outreach**: Notify residents in specific pincodes about upcoming MBU requirements.

#### Data Privacy & Ethics
> **MANDATORY DISCLOSURE:**
> - This analysis uses **anonymized and aggregated** data only.
> - No individual Aadhaar numbers or personally identifiable information (PII) are present in the dataset.
> - No cross-departmental data linkage has been performed.
> - This notebook demonstrates analytics feasibility for government planning purposes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from datetime import timedelta

# Set plot style
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

### 2. Load & Merge Data

We will load the continuation CSV files for Biometric, Demographic, and Enrolment data and concatenate them into unified dataframes.

```python
def load_and_concat(pattern):
    files = glob.glob(pattern)
    df_list = [pd.read_csv(f) for f in files]
    return pd.concat(df_list, ignore_index=True)

# Paths to data folders
base_path = '/home/ubuntu/project_setu_data/data/'

# Load datasets
df_bio = load_and_concat(os.path.join(base_path, 'biometric/*.csv'))
df_demo = load_and_concat(os.path.join(base_path, 'demographic/*.csv'))
df_enrol = load_and_concat(os.path.join(base_path, 'enrollment/*.csv'))

# Convert date to datetime
for df in [df_bio, df_demo, df_enrol]:
    df['date'] = pd.to_datetime(df['date'], dayfirst=True)

print(f"Biometric Data: {df_bio.shape}")
print(f"Demographic Data: {df_demo.shape}")
print(f"Enrolment Data: {df_enrol.shape}")

# Validate Schema
print("\nBiometric Columns:", df_bio.columns.tolist())
print("Enrolment Columns:", df_enrol.columns.tolist())
```

### 3. Construct MBU Demand Signal

The primary signal for MBU demand is `bio_age_5_17`. We also consider enrolment data for younger age groups (`age_0_5`) as leading indicators, as these children will eventually require MBUs when they turn 5.

```python
# Primary demand variable: bio_age_5_17
# We will focus on the biometric update data as it directly represents the MBU activity.
df_demand = df_bio[['date', 'pincode', 'bio_age_5_17']].copy()
df_demand.rename(columns={'bio_age_5_17': 'mbu_demand'}, inplace=True)

print("Demand signal constructed.")
```

### 4. Aggregate by Month & Pincode

To forecast demand, we aggregate the daily data into monthly buckets for each pincode.

```python
# Create month column
df_demand['month'] = df_demand['date'].dt.to_period('M')

# Aggregate by month and pincode
df_monthly = df_demand.groupby(['month', 'pincode'])['mbu_demand'].sum().reset_index()
df_monthly['month'] = df_monthly['month'].dt.to_timestamp()

print(f"Aggregated data: {df_monthly.shape}")
df_monthly.head()
```

### 5. Forecast Next 6 Months

We will select a representative pincode with high activity to demonstrate the forecasting model. We use **Exponential Smoothing (Holt-Winters)**, a classical and explainable time-series model suitable for government planning.

```python
# Select a representative pincode (one with the most data points/highest demand)
target_pincode = df_monthly.groupby('pincode')['mbu_demand'].sum().idxmax()
df_pincode = df_monthly[df_monthly['pincode'] == target_pincode].sort_values('month')
df_pincode.set_index('month', inplace=True)

print(f"Forecasting for Pincode: {target_pincode}")

# Simple Forecasting Model: Exponential Smoothing
model = ExponentialSmoothing(df_pincode['mbu_demand'], trend='add', seasonal=None).fit()
forecast = model.forecast(6)

# Create forecast dataframe
forecast_dates = [df_pincode.index[-1] + pd.DateOffset(months=i) for i in range(1, 7)]
df_forecast = pd.DataFrame({'month': forecast_dates, 'forecast': forecast.values})
df_forecast.set_index('month', inplace=True)

print("6-Month Forecast generated.")
```

### 6. Visualization

Visualizing the historical trends alongside the 6-month forecast to identify potential surges.

```python
plt.figure(figsize=(14, 7))
plt.plot(df_pincode.index, df_pincode['mbu_demand'], label='Historical MBU Demand', marker='o', color='#1f77b4')
plt.plot(df_forecast.index, df_forecast['forecast'], label='Forecasted Demand (Next 6 Months)', linestyle='--', marker='s', color='#ff7f0e')

plt.title(f'MBU Demand Forecast (Age 5-17) for Pincode: {target_pincode}', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Biometric Updates', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### 7. Interpretation & UIDAI Action Plan

#### Interpretation of Results
- **Surge Periods**: The forecast identifies upcoming months with higher-than-average MBU demand. These often correlate with school admission cycles or local awareness drives.
- **Trend Analysis**: A steady increase indicates a growing population of children reaching the age of 5 or 15 in this specific pincode.
